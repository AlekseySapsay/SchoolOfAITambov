Что же такое случайный лес?

Это композиция (ансамбль) из некоторого количества решающих деревьев, каждое из которых построенно на неполном объёме данных, по этому их структура и ответ отличаются. Каждое дерево делает предсказание независиммо от всех остальных, после их ответ усредняется.

Для понимания работы случайного леса, достаточно понимать алгоритм работы одного дерева. Посмотрим на пример, где нам надо классифицировать точки по их цвету. Обучающая выборка выглядит так.

Дерево ищет "лучшую" точку для разбиения на каждом шаге

В конечно итоге

Так выглядит структура разбиения, что для нас, людей, будет означить следующее

Соответственно в момент предсказания, мы просто смотрим куда попадает точка, чтобы предсказать её цвет

С одним деревом разобрались, теперь лес.

Основная идея заключается в том, что при построении каждого дерева из обучающей выборки, предоставленной этому дереву, исключается часть данных (по умолчанию 37%). Соответсвенно каждое дерево получается отличным от других, после построения ансамбля деревьев в момент предсказания, каждое из них "голосует", ответом является среднее.

Эта техника позволяет снизить дисперсию (variance) обучаемого классификатора, уменьшая величину, на сколько ошибка будет отличаться, если обучать модель на разных наборах данных, или другими словами, предотвращает переобучение. Эффективность  достигается благодаря тому, что базовые алгоритмы, обученные по различным подвыборкам, получаются достаточно различными, и их ошибки взаимно компенсируются при голосовании, а также за счёт того, что объекты-выбросы могут не попадать в некоторые обучающие подвыборки.

Строим лес

Голосуем

Стоит ещё раз подчеркнуть, что основной идеей техники является то, что ошибки не коррелируют друг с другом, в отличие от истинных данных, по этому на пересечении деревьев ошибки зануляются (в теории).
